```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
st_df = pd.read_csv("store_data aisiri.csv",header=None)
st_df
```

```python
l = []
for i in range(1,7501):
    l.append([str(st_df.values[i,j]) for j in range(0,20)])
```

```python
l
```

```python
association_rules = apriori(l, min_support=0.001, min_confidence = 0.2 , min_lift = 2 , min_length=2)
association_rules = list(association_rules)
```

```python
association_rules
```

```python
for item in association_rules:
    #first index of the inner list
    #Contains base item and add item
    pair=item[0]
    #23 print (pair)
    #23 print("======")
    items=[x for x in pair]
    #%23 print(items)
    #23 print("=======")
    print("Rule:"+items[0]+"->"+items[1])
    #second index of the inner list
    print("Support: " +str(item[1]))
    #third index of the List Located at eth position
    #of the third index of the inner list
    print("Confidence:"+str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    
```
